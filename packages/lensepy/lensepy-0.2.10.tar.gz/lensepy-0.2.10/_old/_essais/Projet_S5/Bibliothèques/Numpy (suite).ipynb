# Indexation (suite)

```python
import numpy as np

#Cas vecteurs
x_vect = np.array([5,10,15,20,25])

#On peut indexer d'une valeur d'indexe i à j grâce à [i:j]
print("Indexation par [2:5] :", x_vect[2:5])

#Laisser les i et j vides permet de prendre tout du début (comme si i=0) jusqu'au bout (comme si j=len(x_vect))
print("Indexation par [:] :", x_vect[:])


#Cas matrices
#Ce genre de matrice sera celui souvent rencontré lors de la lecture de points dans un fichier CSV
x_mat = np.transpose((np.arange(6),[150,220,100,120,170,160]))
print("Matrice avec deux colonnes et 6 lignes :\n", x_mat)

#Pour le cas des matrices, il suffit d'indexer comme précédemment
val_x = x_mat[:,0] #Valeurs de la première colonne
val_y = x_mat[:,1] #Valeurs de la deuxième colonne

print("Valeurs de la première colonne :\n",val_x)
print("Valeurs de la deuxième colonne :\n",val_y)
```

# Lecture de fichier .csv (tableur excel) et représentation graphique

```python
from matplotlib import pyplot as plt

#On ouvre le fichier csv, numpy créé alors un array avec des colonnes contenant chacunes les données de chaque colonnes du fichier
data = np.loadtxt("solar_spectrum.csv") #np.loadtxt(destination_du_fichier)
print("Données extraites du fichier (spectre du rayonnement solaire) :\n", data)

#On va extraire la première collone (correspondant aux longueurs d'onde) et la deuxième (correspondant à l'irradiance spectrale solaire)

x = data[:,0]
y = data[:,1]

print("Données de la première colonne :\n",x)

#On peut alors représenter dans un graphique ces données grâce à matplotlib

plt.figure(figsize=(15,10))
plt.plot(x,y)
plt.xlabel("λ (nm)")
plt.ylabel("SSI (W.m^{-2}.nm^{-1})")
plt.title("Spectre du rayonnement solaire")
plt.show()

#On remarque que les données vers la fin ne sont pas forcément intéressantes
#On "coupe" donc les colonnes pour mieux voir le spectre visible

x_coupé = data[:1500,0]
y_coupé = data[:1500,1]

plt.figure(figsize=(15,10))
plt.plot(x_coupé,y_coupé)
plt.xlabel("λ (nm)")
plt.ylabel("SSI (W.m^{-2}.nm^{-1})")
plt.title("Zoom sur le spectre du rayonnement solaire")
plt.show()
```
