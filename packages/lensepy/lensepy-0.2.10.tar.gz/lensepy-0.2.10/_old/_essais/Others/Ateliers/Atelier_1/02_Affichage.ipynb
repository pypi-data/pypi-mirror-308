## Affichage de vecteurs

```python
import numpy as np
```

```python
v = np.logspace(1,10,1001)
```

Biblioth√®que Matplotlib / Pyplot

```python
plt.figure()
plt.plot(v)
plt.show()
```

## Fonction y = f(x)

```python
x = np.linspace(0,0.01,1001)
y = np.sin(2*np.pi*400*x)
plt.figure()
plt.title('Sinus 400 Hz')
plt.xlabel('Temps (s)')
plt.ylabel('Amplitude (V)')
plt.plot(x, y)
plt.show()
```

## Afficher une image

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

### Chargement image

```python
image = Image.open('airy_2mm.bmp')
plt.figure()
plt.imshow(image)
plt.show()
print(type(image))
width, height = image.size
print("H = "+str(height)+" / L = "+str(width))
```

### Point sur l'image / Interaction

### Conversion de l'image en matrice

```python
# load the image
image = Image.open('airy_2mm.bmp')
# convert image to numpy array
image_data = np.array(image)
print(type(image_data))
# summarize shape
print(image_data.shape)
```

### Coupe transversale

```python
coupe_hor = image_data[280,:]
plt.figure()
plt.plot(coupe_hor/255)
plt.show()
```

### Maximum d'une image

```python
h_image = image_data.shape[0]
l_image = image_data.shape[1]
print("H = "+str(h_image)+" / L = "+str(l_image))

max_image = np.argmax(image_data)
max_image_iH = np.floor(max_image % h_image)
max_image_iL = np.floor(max_image / h_image)
print(max_image)
print("Ligne = "+str(int(max_image_iL)))
print("Colonne = "+str(int(max_image_iH)))
```
